```python
import polars as pl
```

# 1 Load data

```python
df = pl.read_parquet("./output/berikade_valutor.parquet")
df.tail()
```

```python
df.describe()
```

# 2 Visualising

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))
data = df.filter(pl.col('currency') == 'SEK')
sns.lineplot(data=data,
            x='date',
            y='percent_change').set(title='Daily exchange rate increase')
```

```python
grouped_per_week = df.group_by_dynamic("date", every="1w", by='currency').agg(pl.col(['value', 'percent_change']).mean())
grouped_per_week
```

```python
plt.figure(figsize=(15,8))
currency = "SEK"
sns.lineplot(data=grouped_per_week.filter(pl.col('currency') == currency),
            x='date',
            y='percent_change')
sns.lineplot(data=grouped_per_week.filter(pl.col('currency') == currency).sort(by="date").with_columns(pl.col('percent_change').rolling_mean(window_size="1mo", by="date", closed="right")),
            x='date',
            y='percent_change').set(title='rolling average')
```

# Intermediate visualizations
## Time series

```python
prepped_df = df.with_columns(week = pl.col('date').dt.week(), year= pl.col('date').dt.year(), weekday = pl.col('date').dt.weekday())
df_sek = prepped_df.filter(pl.col("currency") == "SEK")

# Skapa ett eget tema eller använt ett fungerade
sns.set_theme(style="dark")

g = sns.relplot(
    data=df_sek,
    x="week", y="value", col="year", hue="year",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=1, height=3, aspect=2, legend=False,
)

# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=df_sek, x="week", y="value", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::2])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Exchange rate")
g.tight_layout()
plt.savefig("seaborn_plot.png", format='png')

```

## Med distributioner

```python
g = sns.kdeplot(
    data=prepped_df.filter(pl.col('currency').is_in(['SEK','NOK'])),
    x="percent_change", hue="currency",
    fill=True, common_norm=False,
    alpha=.7, linewidth=0,)
```

## Altair exempel

```python
import altair as alt

threshold = 11

bars = alt.Chart(grouped_per_week.filter(pl.col("currency") == "SEK")).mark_bar(color="steelblue").encode(
    x="date:T",
    y="value:Q",
)

highlight = bars.mark_bar(color="#e45755").encode(
    y2=alt.Y2(datum=threshold)
).transform_filter(
    alt.datum.value > threshold
)

rule = alt.Chart().mark_rule().encode(
    y=alt.Y(datum=threshold)
)

label = rule.mark_text(
    x="width",
    dx=-2,
    align="left",
    baseline="bottom",
    text="high exchange"
)

(bars + highlight + rule + label)
```

```python
import duckdb
```

```python
con = duckdb.connect("duck.db")
con.sql("SELECT * FROM prepped_df where percent_change > 4").show()
```

```python
con.sql("""FROM 'output/country_info_iso3666.parquet' SELECT "ISO4217-currency_alphabetic_code", languages """).describe()
```

```python
con.sql("CREATE OR REPLACE TABLE currency_data as SELECT * FROM prepped_df")
con.sql("CREATE OR REPLACE TABLE country as FROM 'output/country_info_iso3666.parquet' SELECT *")
```

```python
con.sql("""
    CREATE OR REPLACE TABLE obt_currency as 
    select 
    currency_data.currency as currency_code, 
    value, 
    percent_change, 
    date, 
    year, 
    week, 
    weekday,
    currency_data."country/currency" as currency_full_name,
    is_independent, 
    "Developed / Developing Countries" as is_developed,
    "Sub-region name" as sub_region,
    "region name" as region,
    "Land Locked Developing Countries (LLDC)" as lldc,
    "Intermediate Region Name" intermediate_region_name,
    official_name_en as official_name,
    "Intermediate Region Name" as region_name,
    continent,
    languages[0] as language    
    from currency_data
    LEFT JOIN country on currency_data.currency = country."ISO4217-currency_alphabetic_code"
    """)
```

```python
con.sql("select * from obt_currency").pl()
```
